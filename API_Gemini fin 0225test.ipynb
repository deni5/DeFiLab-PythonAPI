```python
import google.generativeai as genai
import yfinance as yf

# 🔹 Введи свій API-ключ Gemini
genai.configure(api_key="Your_API_key")

# 🔹 Отримуємо фінансові дані компанії (Apple)
ticker_symbol = "AAPL"
ticker = yf.Ticker(ticker_symbol)
financial_data = ticker.financials.to_json()

# 🔹 Формуємо запит для AI
prompt = f"""
Ти фінансовий аналітик. Проаналізуй фінансові звіти компанії {ticker_symbol} 
та зроби висновки щодо її прибутковості та фінансової стабільності.

Ось фінансові показники:
{financial_data}

Дай короткі висновки щодо поточної фінансової ситуації.
"""

# 🔹 Виконуємо запит до Google Gemini AI
model = genai.GenerativeModel("gemini-pro")
response = model.generate_content(prompt)

# 🔹 Виведення відповіді AI
print(f"🤖 AI аналіз:\n{response.text}")
```
